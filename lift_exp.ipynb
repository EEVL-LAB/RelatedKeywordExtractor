```python
df = pd.read_csv('ex.csv')
```

```python
df = df[['src', 'dst']]
```

```python
d = df.groupby(
    by=['src', 'dst'],
    axis=0
).size()
```

```python
pd.set_option('display.max_columns', None)
```

```python
d
```

```python
n = d.sum()
n
```

```python
d.loc[("예술작품", "이러하다")]
```

```python
d.loc["예술작품"].index.tolist()
```

```python
def support(items):
    d_ = d.loc[tuple(items)]
    if len(items)==1:
        p = d_.sum()
    else:
        p = d_
    r = p/n
    print(items, r)
    return r
    
def lift(items):
    print(
        support(items) / (support([items[0]])*support([items[1]]))
    )

# support(['예술작품', '다양하다'])
lift(['예술작품', '다양하다'])
lift(['예술작품', '듀이'])
lift(['예술작품', '이러하다'])
lift(['예술작품', '위스키'])
lift(['예술작품', '3D스케치'])
```
